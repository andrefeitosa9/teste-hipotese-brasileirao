```python
# Substituir o caractere de traço por hífen para garantir uma separação correta
df['Score'] = df['Score'].str.replace('–', '-')

# Dividir a coluna "score" em duas colunas separadas para gols mandante e visitante
df[['Gols_Mandante', 'Gols_Visitante']] = df['Score'].str.split('-', expand=True).astype(int)

# Calcular pontos para mandante e visitante
def calcular_pontos(row):
    if row['Gols_Mandante'] > row['Gols_Visitante']:
        return pd.Series({'Pontos_Mandante': 3, 'Pontos_Visitante': 0})
    elif row['Gols_Mandante'] < row['Gols_Visitante']:
        return pd.Series({'Pontos_Mandante': 0, 'Pontos_Visitante': 3})
    else:
        return pd.Series({'Pontos_Mandante': 1, 'Pontos_Visitante': 1})

# Aplicar a função para calcular os pontos
df[['Pontos_Mandante', 'Pontos_Visitante']] = df.apply(calcular_pontos, axis=1)
```

```python
times_analisados = ['Botafogo (RJ)', 'Corinthians', 'Santos', 'Cruzeiro']
# Dicionário para armazenar os DataFrames de cada time
jogos_times = {}

# Iterar sobre os times
for time in times_analisados:
    # Filtrar os jogos em que o time é mandante ou visitante
    jogos_time = df[(df['Home Team'] == time) | (df['Away Team'] == time)]
    
    # Salvar o DataFrame do time no dicionário
    jogos_times[time] = jogos_time
```

```python
# Separando os dataframes de cada time

jogos_botafogo = jogos_times['Botafogo (RJ)'].copy()
jogos_corinthians = jogos_times['Corinthians'].copy()
jogos_santos = jogos_times['Santos'].copy()
jogos_cruzeiro = jogos_times['Cruzeiro'].copy()
```

```python
# Converter a coluna "Date" para o formato datetime
jogos_botafogo['Date'] = pd.to_datetime(jogos_botafogo['Date'])

# Definir os intervalos de datas para cada técnico
castro_periodo = (pd.to_datetime('2022-03-28'), pd.to_datetime('2023-06-30'))
cacapa_periodo = (pd.to_datetime('2023-07-01'), pd.to_datetime('2023-07-15'))
lage_periodo = (pd.to_datetime('2023-07-16'), pd.to_datetime('2023-10-03'))
flavio_periodo = (pd.to_datetime('2023-10-04'), pd.to_datetime('2023-11-12'))
nunes_periodo = (pd.to_datetime('2023-11-16'), pd.to_datetime('2024-02-22'))

# Definir uma função para atribuir o técnico com base na data do jogo
def atribuir_tecnico(data):
    if castro_periodo[0] <= data <= castro_periodo[1]:
        return 'Luis Castro'
    elif cacapa_periodo[0] <= data <= cacapa_periodo[1]:
        return 'Claudio Caçapa'
    elif lage_periodo[0] <= data <= lage_periodo[1]:
        return 'Bruno Lage'
    elif flavio_periodo[0] <= data <= flavio_periodo[1]:
        return 'Lucio Flávio'
    elif nunes_periodo[0] <= data <= nunes_periodo[1]:
        return 'Tiago Nunes'
    else:
        return None

# Aplicar a função para atribuir o técnico com base na data do jogo
jogos_botafogo['Técnico'] = jogos_botafogo['Date'].apply(atribuir_tecnico)
```

```python
# Utilizando a coluna de técnicos como referência, tirar a média de pontos feitos nas rodadas de mesmo técnico
# e o desvio padrão da média durante o período do técnico.
# Fazer a mesma coisa com xG, utilizando termos xG A favor e xG Contra
# Salvar um DF com os nomes dos técnicos na primeira coluna e os seguintes dados na primeira linha: Número de jogos desse técnico, pontuacao_media, desvio_padrão_media_pontuação, xG_aFavor_medio, desvio_padrao_xGaFavor, xG_Contra_medio, desvio_padrao_xGaFavor
```

```python
# Corinthians
# Obtendo a ordem original dos técnicos
jogos_corinthians.dropna(how='any', inplace=True)
jogos_corinthians = jogos_corinthians[jogos_corinthians['Técnico'] != 'Fernando Lázaro']
ordem_tecnicos = jogos_corinthians['Técnico'].unique()

# Criando uma categoria ordenada com base na ordem original dos técnicos
categoria_tecnicos = pd.Categorical(jogos_corinthians['Técnico'], categories=ordem_tecnicos, ordered=True)

# Agrupando os dados pelo técnico usando a categoria ordenada
tecnico_agrupado = jogos_corinthians.groupby(categoria_tecnicos)

# Calcular as estatísticas para os jogos em que a equipe é tanto a equipe da casa quanto a equipe visitante
pontos_media = tecnico_agrupado['Pontuação Partida'].mean()
pontos_desvio_padrao = tecnico_agrupado['Pontuação Partida'].std()
xG_afavor_media = tecnico_agrupado['xG_aFavor'].mean()
xG_afavor_desvio_padrao = tecnico_agrupado['xG_aFavor'].std()
xG_contra_media = tecnico_agrupado['xG_Contra'].mean()
xG_contra_desvio_padrao = tecnico_agrupado['xG_Contra'].std()

# Contar o número de jogos para cada técnico
num_jogos = tecnico_agrupado.size()

# Criar um DataFrame com os resultados
resultados = pd.DataFrame({
    'Número de jogos': num_jogos,
    'Pontuação média': pontos_media,
    'Desvio padrão da média de pontos': pontos_desvio_padrao,
    'xG a Favor médio': xG_afavor_media,
    'Desvio padrão do xG a Favor': xG_afavor_desvio_padrao,
    'xG Contra médio': xG_contra_media,
    'Desvio padrão do xG Contra': xG_contra_desvio_padrao
})

# Salvar os resultados em um arquivo CSV
resultados.to_csv('Tabela Médias Treinadores - Corinthians.csv')
```
